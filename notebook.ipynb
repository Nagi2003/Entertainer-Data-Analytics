```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the Excel files
df1 = pd.read_excel("Entertainer - Basic Info.xlsx")
df2 = pd.read_excel("Entertainer - Breakthrough Info.xlsx")
df3 = pd.read_excel("Entertainer - Last work Info.xlsx")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
data = pd.concat([df1 , df2 , df3] , axis = 1 , join='outer')
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
print(data.info())
```

```python
plt.figure(figsize=(20, 5))

# Distribution of Birth Year
plt.subplot(1, 3, 1)
sns.histplot(data['Birth Year'],bins=20,kde=True)
plt.title('Distribution of Birth Year')
```

```python
plt.figure(figsize=(20, 5))
# Distribution of Year of Breakthrough
plt.subplot(1, 3, 2)
sns.histplot(data['Year of Breakthrough/#1 Hit/Award Nomination'],bins=20,kde=True)
plt.title('Distribution of Year of Breakthrough')
```

```python
plt.figure(figsize=(20, 5))
# Distribution of Year of Last Major Work
plt.subplot(1, 3, 3)
sns.histplot(data['Year of Last Major Work (arguable)'],bins=10,kde=True)
plt.title('Distribution of Year of Last Major Work')

plt.tight_layout()
plt.show()
```

```python
# Gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender (traditional)', data=data)
plt.title('Gender Distribution')
plt.show()
```

```python
# Distribution of entertainers still alive vs deceased
plt.figure(figsize=(12, 5))
sns.countplot(data=data, x="Year of Death")
plt.xticks(rotation=45)
plt.title("Alive vs Deceased Entertainers")
plt.show()
```

```python
# Scatter plot of Birth Year vs Year of Breakthrough
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x="Birth Year", y="Year of Breakthrough/#1 Hit/Award Nomination", hue="Gender (traditional)")
plt.title("Birth Year vs Year of Breakthrough")
plt.show()
```
